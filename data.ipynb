HANDLING MISSING VALUES

```python
data.isnull().sum()   #Calculating number of missing values per column
```

```python
import numpy as np

#Filling in missing values with average of corresponding column
def fill_in_values(df_name):
 for column in df_name.select_dtypes(include=[np.number]).columns:
    if df_name[column].isna().sum() > 0:
      avg_val = df_name[column].mean()
      df_name[column] = df_name[column].fillna(avg_val)
fill_in_values(data)
print(data.isna().sum())
```

HANDLING OUTLIERS

```python
#Identifying outliers
def outliers(df_name):
    outlier_dict = {}
    for column in df_name.select_dtypes(include=[np.number]).columns:
       Q1 = df_name[column].quantile(0.25)
       Q3 = df_name[column].quantile(0.75)
       IQR = Q3 - Q1
       lower_bound = Q1 - 1.5*IQR
       upper_bound = Q3 + 1.5*IQR
       outliers = df_name[(df_name[column] < lower_bound) | (df_name[column] > upper_bound)]
       if not outliers.empty:
          outlier_dict[column] = outliers
    return outlier_dict  

  
outliers(data)
 
```

```python
#Handling outliers by removing them

def handle_outliers(df_name):
    
    data_copy = df_name.copy()
    
    numeric_cols = data_copy.select_dtypes(include=[np.number]).columns
    
    # Create a mask for valid values
    mask = pd.Series(True, index=data_copy.index) 
    
    for column in numeric_cols:
        Q1 = data_copy[column].quantile(0.25)
        Q3 = data_copy[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Update the mask to keep only valid values
        mask &= (data_copy[column] >= lower_bound) & (data_copy[column] <= upper_bound)
    
    return data_copy[mask]  # Apply the final mask

df_cleaned = handle_outliers(data)
df_cleaned.describe()
```

DATA VISUALIZATION
